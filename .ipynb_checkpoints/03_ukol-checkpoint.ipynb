### Třetí úkol 
bude pro prodejce doplnit hodnotu ve sloupci web vyextrovanou ze sloupce email pokud ji nemají uvedenou v příslušném formátu www.mujweb.fr. O rozhodnutí doplnění dané adresy prosím použijte četnostní kriterium, pokud je daných webových adres větší než nějaké množství. Např. pro orange.fr tak nemá smysl doplňovat adresu webu neboť se jedná o lokálního zřizovatele schránek podobně jako gmail.com nebo seznam.cz.

```python
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv('dealers.csv')
df['email'] = df['email'].astype('str') 
```

```python
df
```

```python
df_email = df[['web', 'email']]
df_email.head(20)
```

```python
emails = df_email['email'].to_list()
emails

cleansed_emails = []

for email in emails:
    if '@' not in email:
        cleansed_emails.append(email)
        
    else:
        domain = email.split('@')
        cleansed_emails.append(domain[1])
    
cleansed_emails
```

```python
count = Counter(cleansed_emails)
count = count.most_common()
```

```python
count = dict(count)
```

```python
count
```

```python
count.pop('nan')
```

```python
type(count)
```

```python
def extract_domain(email):
    if email != 'nan':
        domain = email.split('@')[1]
        return domain
    return None
```

```python
def should_fill_website(domain, threshold):
    return count.get(domain, 0) < threshold
```

```python
def fill_website(row):
    email = row['email']
    web = row['web']
    if pd.isna(web) and email and email != 'nan':
        domain = extract_domain(email)
        if should_fill_website(domain, threshold=100):
            row['web'] = f'www.{domain}'
    return row
```

```python
df_email = df_email.apply(fill_website, axis=1)
df_email.head(20)
```
