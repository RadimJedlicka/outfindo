### Někteří prodejci se pak mohou ve zdroji dealers.csv vyskytnout vícekrát. 
Kvůli tomu, že např. nabízejí zboží od více výrobců např. (ktm, ghost, scott, cannodale), vaším úkolem bude rozšířit data o jeden sloupec chain, ve kterém bude pro stejného prodejce prodávající více značek stejná hodnota. Pokud se nevyskytují duplicity zůstane sloupec chain nevyplněný.

```python
import pandas as pd
```

```python
df = pd.read_csv('dealers.csv')
df['email'] = df['email'].astype('str') 
df
```

```python
df.groupby(['eshop'], sort = True).count()

# df.groupby(['Name'], sort = False).sum()
```

```python
# import pandas as pd

# data = {
#     'Dealer': ['Dealer A', 'Dealer B', 'Dealer A', 'Dealer C', 'Dealer B'],
#     'Brand': ['KTM', 'Ghost', 'Scott', 'Cannodale', 'KTM']
# }

# df = pd.DataFrame(data)

def add_chain(df):
    dealer_counts = df.groupby('eshop')['brand'].nunique().reset_index()
    duplicated_dealers = dealer_counts[dealer_counts['brand'] > 1]['eshop']

    df['Chain'] = df['eshop'].apply(lambda x: x if x in duplicated_dealers.values else '')

    return df

df = add_chain(df)

df
```
